```python
# Get the document from this Colab Notebook path for extraction: /content/mu_earnings_call.pdf
from pypdf import PdfReader

reader = PdfReader("/content/mu_earnings_call.pdf")
pdf_texts = [p.extract_text().strip() for p in reader.pages]

# Filter the empty strings
pdf_texts = [text for text in pdf_texts if text]

print(pdf_texts[0])
```

```python
#Run the langchain package to split the document text into chunks
from langchain.text_splitter import RecursiveCharacterTextSplitter, SentenceTransformersTokenTextSplitter

character_splitter = RecursiveCharacterTextSplitter(
    separators=["\n\n", "\n", ". ", " ", ""],
    chunk_size=1000,
    chunk_overlap=0
)
character_split_texts = character_splitter.split_text('\n\n'.join(pdf_texts))

print((character_split_texts[5]))
print(f"\nTotal chunks: {len(character_split_texts)}")
```

```python
#Get the tokens for the text chunks
token_splitter = SentenceTransformersTokenTextSplitter(chunk_overlap=0, tokens_per_chunk=256)

token_split_texts = []
for text in character_split_texts:
    token_split_texts += token_splitter.split_text(text)

#print((token_split_texts[5]))
print(f"\nTotal chunks: {len(token_split_texts)}")
```

```python
#Implement ChromaDB for getting the embeddings to store as  the collection
import chromadb
from chromadb.utils.embedding_functions import SentenceTransformerEmbeddingFunction

embedding_function = SentenceTransformerEmbeddingFunction()
#print(embedding_function([token_split_texts[5]]))

chroma_client = chromadb.Client()
chroma_collection = chroma_client.create_collection("micron_earnigns_call", embedding_function=embedding_function)

ids = [str(i) for i in range(len(token_split_texts))]

chroma_collection.add(ids=ids, documents=token_split_texts)
chroma_collection.count()
```

```python
# Implement the LLM to query the document


from langchain_mistralai import MistralAIEmbeddings
from langchain_mistralai.chat_models import ChatMistralAI

from getpass import getpass

import os
from mistralai import Mistral

```

```python
#Test a query to return records for the embeddings  stored in ChromaDB

query = "What was the last quarter's revenue amount?"

results = chroma_collection.query(query_texts=[query], n_results=3)
retrieved_documents = results['documents'][0]

for document in retrieved_documents:
    print((document))
    print('\n')
```

```python
# Define the LLM Client credentials
client = Mistral(api_key="GET_api_from_Mistral_AI")
```

```python
def rag(query, retrieved_documents, llm_client, model="mistral-large-latest"):
    information = "\n\n".join(retrieved_documents)

    messages = [
        {
            "role": "system",
            "content": "You are a helpful expert financial research assistant. Your users are asking questions about information contained in a earnings report."
            "You will be shown the user's question, and the relevant information from the earnings report. Answer the user's question using only this information."
        },
        {"role": "user", "content": f"Question: {query}. \n Information: {information}"}
    ]

    response = llm_client.chat.complete(
        model=model,
        messages=messages,
    )
    content = response.choices[0].message.content
    return content
```

```python
#Define the Prompt based on the VectorDB test query for the records
Prompt = "Provide a summary with a data table. The data table should summarize any dollar revenue growth year over the prior year (2023) or percent change in revenue  compared to last year for major divisions. Assess the guidance provided for any  expected dollar or percentage revenue growth for the next quarter or year. Give the transcript provided a score out of 100 for its sentiment – 100 for perfectly positive sentiment - 50 for neutral - 0 for perfectly negative sentiment. Explain this sentiment score in one sentence after the text summary."
```

```python
#Create the dataframe to store the queries and responses
import pandas as pd

llm_df = pd.DataFrame([Prompt], columns=['LLM query'])
print (llm_df)
```

```python
output = rag(query=Prompt, llm_client = client, retrieved_documents=retrieved_documents)

print((output))
```

```python
import datetime

llm_df['LLM response'] = output
llm_df['LLM tested'] = 'MistralAI'
llm_df['Report Date'] = pd.to_datetime('2024-10-10')


llm_df.head()
```
